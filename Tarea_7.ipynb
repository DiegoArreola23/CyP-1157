# PROBLEMA 3.1

```python
PROPAR = 0
SUMPAR = 0
SUMIMP = 0
CUEPAR = 0
i = 1
while i <= 270:
    NUM = int(input("Dame un nÃºmero entero: ")) 
    if NUM > 0:
        if ((-1)**NUM) > 0:
            SUMPAR = SUMPAR + NUM
            CUEPAR = CUEPAR + 1
            i = i + 1            
        else:
            SUMIMP = SUMIMP + NUM
            i = i + 1
    else:
        i = i + 1
        print(i)
else:
    PROPAR = SUMPAR/CUEPAR
    print(f"PROPAR = {PROPAR} , SIMIMP = {SUMIMP}")        
     
print("SUMPAR = ",SUMPAR)
print("CUEPAR = ",CUEPAR)
print("SUMIMP = ",SUMIMP)
print("FIN DEL PROGRAMA")
```

# PROBLEMA 3.3

```python
SERIE = 0
N = int(input("Dame un valor entero: "))
BAND = 'T'
I = 1
print("N =",N)
print("BAND = ", BAND)
print("SERIE = ",SERIE)
print("------------------------------------------------------------------")
while I <= N:

        if BAND == 'T':
            SERIE = SERIE + (1/I)
            BAND = 'F'
            print("BAND = ", BAND)
            print("SERIE = ",SERIE)
            print("------------------------------------------------------------------")

        else:
            SERIE = SERIE - (1/I)
            BAND = 'T'

            print("BAND = ", BAND)
            print("SERIE = ",SERIE)
            print("------------------------------------------------------------------")
        I = I + 1

print("FIN DEL PROGRAMA")
```

```python
NOM = 0
NSUE = 0
print("NOM = ",NOM)
SUE = float(input("Dame un valor de tipo real: "))
while SUE > -1:
    if SUE < 1000:
        NSUE = SUE*(1.15)
    else:
        NSUE = SUE*(1.12)
    NOM = NOM + NSUE
    print("-----------------------------------------")
    print("NOM = ",NOM)
    SUE = float(input("Dame un valor de tipo real: "))
    print("NSUE = ",NSUE)

print("-----------------------------------------")
print(f"NOM = {NOM}")
print("FIN DEL PROGRAMA")
```
